```python
import pandas as pd
import numpy as np
import re
import emoji
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
from nltk.corpus import stopwords
from sentence_transformers import SentenceTransformer
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import MultinomialNB
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
from sklearn.preprocessing import LabelEncoder, label_binarize, StandardScaler
from imblearn.over_sampling import SMOTE
import warnings
warnings.filterwarnings("ignore")

nltk.download('stopwords')

# Load Data
def load_data():
    df = pd.read_csv("/content/bank dastaset.csv", names=["Sentiment", "News_Headline"], encoding="ISO-8859-1")
    print(df['Sentiment'].value_counts())
    print(df.head())
    return df

# Class Distribution Plot
def plot_class_distribution(df):
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x='Sentiment', palette='Set2')
    plt.title("Class Distribution", fontsize=14, fontweight='bold')
    plt.xlabel("Sentiment")
    plt.ylabel("Count")
    plt.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.show()
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in text.split() if word not in stop_words and len(word) > 2]
    return " ".join(tokens)

# Preprocess
def preprocess_data(df):
    df['Cleaned_Text'] = df['News_Headline'].astype(str).apply(clean_text)
    df = df[df['Cleaned_Text'].str.strip().astype(bool)]
    df = df.reset_index(drop=True)
    return df

# Sentence-BERT Encoding
def encode_with_bert(texts, model_name='all-MiniLM-L6-v2'):
    print("Generating Sentence-BERT embeddings...")
    model = SentenceTransformer(model_name)
    embeddings = model.encode(texts, show_progress_bar=True)
    return np.array(embeddings)

# Train-Test Split and Encoding
def prepare_data(df, embeddings):
    le = LabelEncoder()
    y = le.fit_transform(df['Sentiment'])
    print(f"\nEncoded Class Mapping: {dict(zip(le.classes_, le.transform(le.classes_)))}")

    X_train, X_test, y_train, y_test = train_test_split(
        embeddings, y, test_size=0.2, stratify=y, random_state=42)

    return X_train, X_test, y_train, y_test, le

# Resampling
def resample_data(X_train, y_train):
    smote = SMOTE(random_state=42)
    X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

    print("\nResampled Class Distribution:")
    unique, counts = np.unique(y_resampled, return_counts=True)
    for label, count in zip(unique, counts):
        print(f"Class {label} -> {count} samples")

    return X_resampled, y_resampled

# Model Evaluation
def evaluate_model(model, X_test, y_test, label_encoder):
    y_pred = model.predict(X_test)

    print("\nClassification Report:")
    print(classification_report(y_test, y_pred, target_names=label_encoder.classes_))

    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d',
                xticklabels=label_encoder.classes_,
                yticklabels=label_encoder.classes_,
                cmap='Blues')
    plt.title("Confusion Matrix")
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.tight_layout()
    plt.show()

    # ROC-AUC Score
    y_test_bin = label_binarize(y_test, classes=np.unique(y_test))
    y_pred_prob = model.predict_proba(X_test) if hasattr(model, "predict_proba") else None
    if y_pred_prob is not None:
        roc_auc = roc_auc_score(y_test_bin, y_pred_prob, average="macro", multi_class="ovr")
        print(f"Macro-averaged ROC-AUC Score: {roc_auc:.3f}")

    scores = cross_val_score(model, X_test, y_test, cv=5, scoring='f1_weighted')
    print(f"Cross-validated F1-score (on test set): {scores.mean():.3f} ± {scores.std():.3f}")

# Hyperparameter Tuning
def tune_model(model, param_grid, X_train, y_train):
    grid_search = GridSearchCV(model, param_grid, cv=3, scoring='f1_weighted', verbose=1, n_jobs=-1)
    grid_search.fit(X_train, y_train)
    print(f"\nBest Parameters: {grid_search.best_params_}")
    return grid_search.best_estimator_

# Models and Param Grids
def get_models_with_params():
    return {
        "Logistic Regression": (
            LogisticRegression(max_iter=1000, class_weight='balanced'),
            {
                'C': [0.01, 0.1, 1, 10],
                'solver': ['liblinear', 'lbfgs']
            }
        ),
        "SVM (Linear Kernel)": (
            SVC(kernel='linear', probability=True, class_weight='balanced'),
            {
                'C': [0.1, 1, 10]
            }
        ),
        "Random Forest": (
            RandomForestClassifier(class_weight='balanced'),
            {
                'n_estimators': [100, 200],
                'max_depth': [10, 20, None]
            }
        ),
        "XGBoost": (
            XGBClassifier(use_label_encoder=False, eval_metric='mlogloss'),
            {
                'n_estimators': [100, 200],
                'learning_rate': [0.05, 0.1, 0.2],
                'max_depth': [3, 6, 10]
            }
        ),
        "Gradient Boosting": (
            GradientBoostingClassifier(),
            {}  # Optional: You can add grid if needed
        ),
        "Naive Bayes (replaced with Logistic Regression)": (
            LogisticRegression(max_iter=1000),
            {}  # Using LR instead of NB due to embedding format
        )
    }

# Main Pipeline
def run_pipeline():
    df = load_data()
    plot_class_distribution(df)
    df = preprocess_data(df)
    embeddings = encode_with_bert(df['Cleaned_Text'])
    X_train, X_test, y_train, y_test, le = prepare_data(df, embeddings)
    X_train_res, y_train_res = resample_data(X_train, y_train)

    models_with_params = get_models_with_params()

    for name, (model, param_grid) in models_with_params.items():
        print(f"\n{'='*50}\nTuning & Training: {name}")
        if param_grid:
            model = tune_model(model, param_grid, X_train_res, y_train_res)
        else:
            model.fit(X_train_res, y_train_res)
        evaluate_model(model, X_test, y_test, le)

# Run everything
if __name__ == "__main__":
    run_pipeline()
```
